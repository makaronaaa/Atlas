```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("/content/sample_data/Export (All prevalence covariates).csv")
df.head()
```

```python
age_group_data_all = df[df['Covariate name'].str.contains("age group", na=False)]

if not age_group_data_all.empty:
    age_group_data_sorted_all = age_group_data_all.sort_values(by='Covariate short name')  # Sort for better visualization
    x_labels_age_all = age_group_data_sorted_all['Covariate short name']
    counts_age_all = age_group_data_sorted_all['Count']
    percentages_age_all = age_group_data_sorted_all['Percent']

    # Plotting the age group prevalence distribution with percentages
    plt.figure(figsize=(14, 6))
    bars = plt.bar(x_labels_age_all, counts_age_all, color='cornflowerblue', edgecolor='black')

    for bar, percentage in zip(bars, percentages_age_all):
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
                 f'{percentage:.1f}%', ha='center', va='bottom', fontsize=10)

    plt.title('Prevalence Distribution Across All Age Groups (Counts and Percentages)', fontsize=16)
    plt.xlabel('Age Group', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print("No age group-related data found in the dataset.")
```
